
Let's do a similar thing for the x-axis of your world development chart, with the xticks() function. The tick values 1000, 10000 and 100000 should be replaced by 1k, 10k and 100k. To this end, two lists have already been created for you: tick_val and tick_lab.

Right now, the scatter plot is just a cloud of blue dots, indistinguishable from each other. Let's change this. Wouldn't it be nice if the size of the dots corresponds to the population?

To accomplish this, we will use the list population. It contains population numbers for each country expressed in millions. You can see that this list is added to the scatter method, as the argument s, for size.


```python
# Basic scatter plot, log scale
plt.scatter(df.gdp_cap, df.life_exp, s = df.population * 2)
plt.xscale('log')

# Add labels
plt.xlabel("GDP per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development - 2007")

# Adapt the ticks on the x-axis
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
# add grid
plt.grid()

# After customizing, display the plot
plt.show()
```

The next step is making the plot more colorful. To do this, a list col has been created. It's a list from the DataFrame's population with a color for each corresponding country, depending on the continent the country is part of.

How was the list made? The Gapminder data contains a list continent with the continent each country belongs to. A function was created that maps continents onto colors:

Next steps:

<ul><li>Add c = col to the arguments of the plt.scatter() function.
<li>Change the opacity of the bubbles by setting the alpha argument to 0.8 inside plt.scatter().
<li>Alpha can be set from zero to one, where zero is totally transparent, and one is not at all transparent.

```python
df["continent"].value_counts()
```

there are five types of continent,so we shoud used five separate colors for this continent. 
<ul><li>Africa -> blue
<li>Asia -> red
<li>Europe -> green
<li> Americas -> yellow
<li>Oceania -> black</li></ul>
<p>
then we create a function to add the colours for the relevant continent, next add the new column for Data Frame as color.

```python
# Create function that maps a country's continent to a color
def color_map(c):
    if c["continent"] == 'Africa':
        return 'DARKTURQUOISE'
    
    elif c["continent"] == 'Asia':
        return 'red'
    
    elif c["continent"] == 'Europe':
        return 'LIMEGREEN'
    
    elif c["continent"] == 'Americas':
        return 'yellow'
    
    else:
        return 'black'
```

```python
df['color'] = df.apply(color_map, axis=1)
df.head()
```

```python
df.tail()
```

```python
# Basic scatter plot, log scale
plt.scatter(df.gdp_cap, df.life_exp, s = df.population * 2, c = df.color, alpha = 0.8)
plt.xscale('log')

# Add labels
plt.xlabel("GDP per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development - 2007")

# Adapt the ticks on the x-axis
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
# add grid
plt.grid()

# After customizing, display the plot
plt.show()
```

Now, let's add a legend so people know what the colors indicate.


```python
# Basic scatter plot, log scale
plt.scatter(df.gdp_cap, df.life_exp, s = df.population * 2, c = df.color, alpha = 0.8)

# Add label
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development - 2007")

# Adapt the ticks on the x-axis
plt.xticks([1000, 10000,100000],['1k', '10k', '100k'])

# add grid
plt.grid(True)

# Add the Legend
colors = ["DARKTURQUOISE", "red", "LIMEGREEN", "yellow","black"] 
labels = ["Africa", "Asia", "Europe", "Americas", "Oceania"]
patches = [mpatches.Patch(color = colors[i], label = "{:s}".format(labels[i])) for i in range(len(labels))]
plt.legend(handles = patches, bbox_to_anchor=(0.28, 0.6))

# After customizing, display the plot
plt.show()
```

Next we will highlight China and India on the plot and the plot is complete and ready to save the figure and submit for publishing.

```python
# Basic scatter plot, log scale
plt.scatter(df.gdp_cap, df.life_exp, s = df.population * 3, c = df.color, edgecolors="black", linewidth=0.8, alpha = 0.9)

# Add label
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development - 2007")

# Adapt the ticks on the x-axis
plt.xticks([1000, 10000,100000],['1k', '10k', '100k'])

# add grid
plt.grid(True)

# Add the Legend
colors = ["DARKTURQUOISE", "Crimson", "LIMEGREEN", "yellow","black"] 
labels = ["Africa", "Asia", "Europe", "Americas", "Oceania"]
patches = [mpatches.Patch(color = colors[i], label = "{:s}".format(labels[i])) for i in range(len(labels))]
plt.legend(handles = patches, bbox_to_anchor = (0.209, 0.99),title='colour by continent')


# Additional customizations to highlight China and Indi
for i in range(len(df)):
    if df.population[i] < 2:
        plt.text(df.gdp_cap[i], df.life_exp[i], df.country[i], fontsize= df.population[i] * 3,fontname='Times New Roman')
    elif df.population[i] < 5:
        plt.text(df.gdp_cap[i], df.life_exp[i], df.country[i], fontsize= df.population[i] * 2.09,fontname='Times New Roman')
    else:
        plt.text(df.gdp_cap[i], df.life_exp[i], df.country[i], fontsize= df.population[i] * 0.0155, fontname='Times New Roman')
    
fig = plt.gcf()
fig.set_size_inches(10, 8)


# After customizing, display the plot
plt.show()
```
